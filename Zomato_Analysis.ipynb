# Removing 'New' , '-' and '/5' From rate

```python
def handlerate(value):
    if (value== 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
zomato['rate'] = zomato.rate.apply(handlerate)    
zomato['rate'].head()
```

```python
zomato.rate.isnull().sum()
```

# Filling Null Values in Rate Column With Mean

```python
zomato['rate'].fillna(zomato.rate.mean() , inplace = True )
zomato.rate.isnull().sum()
```

```python
zomato.info()
```

```python
zomato.dropna(inplace = True)
zomato.head()
```

# Listed in(city) and Location , both are there , lets keep only one

```python
df = df.drop(['listed_in(city)'], axis= 1)
```

```python
df.Cost2plates.unique()
```

# Removing , from Cost2Plates Column

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
```

```python
df['Cost2plates'] = df.Cost2plates.apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

# Cleaning Rest Type Column

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types # restorants type
```

```python
rest_type_lessthan_1000 = rest_types[rest_types < 1000]
rest_type_lessthan_1000
```

# Making Rest Types less than 1000 in frequency as others

```python
def handle_rest_types(value):
    if (value in rest_type_lessthan_1000):
        return 'others'
    else:
        return value
    
```

```python
df['rest_type'] = df['rest_type'].apply(handle_rest_types)
df['rest_type'].value_counts()
```

# Cleaning Location Column

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = False)

location_lessthan_300 = location[location < 300]

def handle_location(value):
    if(value in location_lessthan_300):
        return 'others'
    else:
        return value
df['location'] = df['location'].apply(handle_location)    
df.location.value_counts()
```

```python
df.info()
```

```python
df.head()
```

# Cleaning Cuisines Column

```python
df.cuisines.value_counts()
```

```python
cuisines = df.cuisines.value_counts(ascending = True)
cuisines_lessthan_100 = cuisines[cuisines < 100]

def handle_cuisines(value):
    if value in cuisines_lessthan_100:
        return 'others'
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(handle_cuisines) 
df['cuisines'].value_counts()
```
